Judging from above, it seems as if OPS and WAR two heavily weighted statistics used by the BBWAA in voting(for position players)

What statistics hold the most weight for pitchers?:

Sort by Wins and Innings Pitched(the higher, the better):

```python
df.sort_values(['IP:','W:'], ascending=False)
```

WHIP is the average walks and hits allowed per inning pitched. ERA is the average earned runs allowed per 9 innings pitched(the lower, the better)

```python
df.sort_values(['WHIP:','ERA:'])
```

Judging from above, WHIP and ERA seem to have more wegiht than Wins and Innings Pitched
